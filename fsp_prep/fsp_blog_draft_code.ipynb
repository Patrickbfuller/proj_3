```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
data, sr = librosa.load('/Users/patrickfuller/Desktop/Truth Dare copy.mp3', sr=None)
data.shape, sr
```

```python
spec = librosa.feature.melspectrogram(y=data, sr=sr)
```

```python
import librosa.display
```

```python
librosa.display.specshow(spec,y_axis='mel', x_axis='s', sr=sr)
plt.colorbar();
```

```python
db_spec = librosa.power_to_db(spec, ref=np.max,)
```

```python
librosa.display.specshow(db_spec,y_axis='mel', x_axis='s', sr=sr)
plt.colorbar();
```

```python
librosa.get_duration(data,sr=sr)
```

```python
data_harm, data_perc = librosa.effects.hpss(data)

spec_harm = librosa.feature.melspectrogram(data_harm, sr=sr)
spec_perc = librosa.feature.melspectrogram(data_perc, sr=sr)
db_spec_harm = librosa.power_to_db(spec_harm, ref=np.max)
db_spec_perc = librosa.power_to_db(spec_perc, ref=np.max)
```

```python
librosa.display.specshow(db_spec_harm,y_axis='mel', x_axis='s', sr=sr)
plt.title('Percussive Spectrogram')
plt.colorbar();
```

```python
librosa.display.specshow(db_spec_perc,y_axis='mel', x_axis='s', sr=sr)
plt.title('Harmonic Spectrogram')
plt.colorbar();
```

```python
chroma = librosa.feature.chroma_cqt(y=data_harm, sr=sr)
```

```python
plt.figure(figsize=(18,5))
librosa.display.specshow(chroma, sr=sr, x_axis='time', y_axis='chroma', vmin=0, vmax=1)

plt.title('Chromagram')
plt.colorbar()
```
