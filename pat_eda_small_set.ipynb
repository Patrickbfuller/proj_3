```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import numpy as np
plt.style.use('ggplot')
```

```python
df = pd.read_csv('data.csv', index_col='Unnamed: 0')
```

```python
df.head(2)
```

```python
elec = df[df.label=='Electronic']
metal = df[df.label=='Metal']
hip_hop = df[df.label=='Hip_Hop']
jazz = df[df.label=='Jazz']
classical = df[df.label=='Classical']
country = df[df.label=='Country']
plt.hist(elec.mfcc_1, alpha=0.5)
plt.hist(hip_hop.mfcc_1, alpha=0.5)
```

```python
col = 'chroma_stft'

for g1, g2 in itertools.combinations(df.label.unique(), 2):
    plt.figure(figsize=(6,3))
    plt.hist(df.loc[df.label == g1, col], alpha=0.5, label=g1)
    plt.hist(df.loc[df.label == g2, col], alpha=0.5, label=g2)
    plt.title(col)
    plt.legend()
    plt.show()
```
