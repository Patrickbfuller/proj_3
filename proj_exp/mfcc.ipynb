```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
data, sr = librosa.load('/Users/patrickfuller/Desktop/Lock & Key Dare+Mix copy.mp3', sr=None)
```

```python
spec = librosa.feature.melspectrogram(data, sr=sr)
db_spec = librosa.power_to_db(spec,ref=np.max)
```

```python
plt.figure()
librosa.display.specshow(db_spec, sr=sr, y_axis='mel', x_axis='time')
plt.ylim(0,2000)
```
