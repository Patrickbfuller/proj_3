```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
plt.style.use('seaborn')
```

```python
df = pd.read_csv('data_3k.csv')
print(df.shape)
df.head(3)
```

```python
col_1 = 'chroma_stft'
col_2 = 'mfcc_2'
col_3 = 'mfcc_11'
cols = [col_1, col_2, col_3]
for g1, g2 in itertools.combinations(df.label.unique(), 2):
    plt.figure(figsize=(15,2))
    for i, col in enumerate(cols):
        plt.subplot(1,3,i+1)
        plt.hist(df.loc[df.label == g1, col], alpha=0.5, label=g1)
        plt.hist(df.loc[df.label == g2, col], alpha=0.5, label=g2)
        plt.title(col)
        plt.legend()
    plt.show()
```
